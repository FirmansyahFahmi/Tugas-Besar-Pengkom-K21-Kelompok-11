```python
list_percentile = []
list_percentile.append(["Persen", "Tahun Pendirian", "Daya Tampung", "Luas Bangunan", "Luas Tanah"])

for num in range(10, 110, 10):
    percent = round(num * 0.01, 1)
    start_percent = round(percent - 0.1, 1)
    end_percent = percent
    str_percentage = f"{(int(start_percent * 100))}% - {int(end_percent * 100)}%"
    str_nominal1 = f"{int((data.quantile(end_percent)['tahun_pendirian']))}"
    str_nominal2 = f"{int((data.quantile(end_percent)['daya_tampung']))}"
    str_nominal3 = f"{int((data.quantile(end_percent)['luas_bangunan']))}"
    str_nominal4 = f"{int((data.quantile(end_percent)['luas_tanah']))}"
    list_percentile.append([str_percentage,str_nominal1,str_nominal2,str_nominal3,str_nominal4])

fig, ax = plt.subplots()
table = ax.table(cellText=list_percentile, loc='center')
table.set_fontsize(10)
table.scale(1,2)
ax.axis('off')
plt.title("Persentile Tahun Pendirian, Daya Tampung, Luas Bangunan, dan Luas Tanah")
plt.show()
```

```python
a = data['kecamatan'].unique()
list_kecamatan = []
list_kecamatan.append(["Kecamatan", "Tempat Ibadah"])

for i in a:
    n = data['kecamatan'].value_counts()[i]
    list_kecamatan.append([i, n])

fig, ax = plt.subplots()
table = ax.table(cellText=list_kecamatan, loc='center')
table.set_fontsize(10)
table.scale(1,2)
ax.axis('off')
plt.title("Distribusi Tempat Ibadah berdasarkan Kecamatan")
plt.show()
```

```python
maxim_tahun = data['tahun_pendirian'].max()
minim_tahun = data['tahun_pendirian'].min()
range_tahun = maxim_tahun-minim_tahun
2007, 1975, 1940, 1905, 1870, 1835, 1800, 1765, 1730, 1695, 1650
delta_tahun = [1695, 1870, 1905, 1940, 1975, 2007]
freq = 0
prev_tahun = 1649
list_frek = []
list_frek.append(["Range Tahun", "Tempat Ibadah"])

for i in delta_tahun:
    end_range = i
    start_range = prev_tahun
    query =  (start_range <= data['tahun_pendirian']) & (data['tahun_pendirian'] <= end_range)  

    data1 = data.loc[query].shape[0]
    freq += data1
    str_frek = f"{start_range+1} - {end_range}"
    str_nominal = f"{data1}"
    list_frek.append([str_frek,str_nominal])
    prev_tahun = end_range

fig, ax = plt.subplots()
table = ax.table(cellText=list_frek, loc='center')
table.set_fontsize(12)
table.scale(1,2)
ax.axis('off')
plt.title("Distribusi Tempat Ibadah berdasarkan Tahun")
plt.show()
```
